## Let's revisit our guessing game

Remember this?

```python
import random
secret_number = random.randint(1,100)
guess = 0
guess_counter = 0

while guess != secret_number:
    guess = int(input("Guess a number: "))
    guess_counter += 1 #this is the same as guess_counter = guess_counter + 1

    if guess == secret_number:
        print("That was right, good guess!")
    elif guess < secret_number:
        print("Try a little higher.")
    elif guess > secret_number:
        print("Try a little lower.")
    else:
        print("Wrong!")
        
print("That took",guess_counter,"guesses")
```
